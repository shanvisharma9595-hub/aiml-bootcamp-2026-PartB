```python
#6.
students = [
    {"Student_name":"Aarav", "marks":82, "attendance":88, "project_completed":True},
    {"Student_name":"Riya", "marks":68, "attendance":55, "project_completed":False},
    {"Student_name":"Karan", "marks":94, "attendance":92, "project_completed":True},
    {"Student_name":"Meera", "marks":61, "attendance":70, "project_completed":False},
    {"Student_name":"Arjun", "marks":97, "attendance":96, "project_completed":True}
]

for student in students:
    print(student["Student_name"])

    if student["marks"] > 75:
        print("Distinction")

    if student["attendance"] < 75:
        print("Low Attendance")

    if student["project_completed"] == False:
        print("Project Not Completed")

    print()
```
